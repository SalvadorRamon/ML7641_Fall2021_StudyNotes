```python
from PIL import Image
import mnist
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
```

```python
# training
X_train = mnist.train_images()
y_train = mnist.train_labels()

X_test = mnist.test_images()
y_test = mnist.test_labels()

#print("X train", X_train)
#print("y train", y_train)
#print("X test", X_test)
#print("y test", y_test)

# dimension
print(X_train.ndim)

#shape
print(X_train.shape)
```

```python
X_train = X_train.reshape((-1, 28*28))
X_test = X_test.reshape((-1, 28*28))

#shape
print(X_train.shape)
```

```python
# if we look at the 1st element of x train.
# The number is not ideal.
# We want to make it into range of 0-1

print(X_train[0])

print("scale down to optimize")

X_train = (X_train/256)
X_test = (X_test/256)

print(X_train[0])
```

```python
# model
clf = MLPClassifier(solver = 'adam', activation = 'relu', hidden_layer_sizes = (64, 64))
clf.fit(X_train, y_train)
```

```python
prediction = clf.predict(X_test)
acc = confusion_matrix(y_test, prediction)
```

```python
print("acc", acc)
```

```python
def accuracy(confusion_matrix):
    diagonal = confusion_matrix.trace()
    elements = confusion_matrix.sum()
    return diagonal/elements
```

```python
print(accuracy(acc))
```
